```python
import os
import sys

project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.append(project_root)
```

```python
GENERATED_SAMPLE_DATA_DIR = 'generated_sample_data'
OUTPUT_FILE_NAME = 'summarized_reviews.csv'
```

```python
#if you need an API Key from OpenAI
#https://platform.openai.com/account/api-keys

from openai import OpenAI
from dotenv import load_dotenv, find_dotenv

_ = load_dotenv(find_dotenv(), override=True)

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
```

```python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)

prompt = f"""
    You're an assistant that summarizes product reviews.
    Summarize the user review in about 2 or 3 sentences, \
    clearly and concisely, in english.
"""

system_prompt = { 'role': 'system', 'content': prompt }

def get_summarized_review(review, temperature=0):
    review_prompt = { 'role': 'user', 'content': review }
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[system_prompt, review_prompt],
        temperature=temperature,
    )

    return response.choices[0].message.content
```

```python
from lib.data_load import load_data

df = await load_data()
print(f"\n{'CONCATENATED':<62} {df.shape[0]:>6,} rows x {df.shape[1]} cols")
df.sample(10)
```

```python
from lib.data_products_clean import clean_data_frame

df = clean_data_frame(df)
print(df.isna().sum())
display(df.sample(10))
```

```python
df['reviews.content'] = df['reviews.title'] + " " + df['reviews.text']
df['reviews.summary'] = df['reviews.content'].apply(get_summarized_review)
display(df.sample(10))
```

```python
# Save summarized reviews into CSV
df.to_csv(os.path.join(GENERATED_SAMPLE_DATA_DIR, OUTPUT_FILE_NAME), index=False)
```
