### Extract random SEED data for Django localhost database

```python
import os
import sys
import pandas as pd

project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.append(project_root)
```

```python
OUTPUT_SEED_DATA_DIR = '../storage/seed_data'
OUTPUT_FILE_NAME = 'products.csv'
```

```python
from lib.data_load import load_data

df = await load_data()
print(f"\n{'CONCATENATED':<62} {df.shape[0]:>6,} rows x {df.shape[1]} cols")
df.sample(10)
```

```python
df_output = pd.DataFrame()
random_products = df.sample(5)
for i, row in random_products.iterrows():
    row_index = df.index.get_loc(i)
    df_output = pd.concat([df_output, df.iloc[row_index:row_index + 5]]).reset_index(drop=True)
    
print(df_output['asins'].value_counts())

df_output.to_csv(os.path.join(OUTPUT_SEED_DATA_DIR, OUTPUT_FILE_NAME), index=False)
```
